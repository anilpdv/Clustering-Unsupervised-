```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
X = df.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X,method='ward'))
#plt.figure(figsize=(12,9))
plt.title('Dendogram')
plt.xlabel('customers')
plt.ylabel('Eucleaden distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
plt.style.use('classic')
plt.figure(figsize=(12,9))
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=200,c='red',label='clus1')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=200,c='blue',label='clus2')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=200,c='green',label='clus3')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=200,c='cyan',label='clus4')
plt.scatter(X[y_hc==4,0],X[y_hc==4,1],s=200,c='violet',label='sensible')
#plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='centroids')
plt.title('Cluster of clients')
plt.xlabel('Annual Income ')
plt.ylabel('spending scores')
plt.legend()
plt.show()
```
